```python
# Import necessary packages

import pandas as pd
import os
import shutil

```

```python
# Create the data for positive samples
# First parser the file which is availble in chestxray dataset.. file name is metadata.csv

FILE_PATH = "chestxray/metadata.csv"
# parser the image folder

IMAGES_PATH ="chestxray/images" 
```

```python
# read the metadat file using pandas
df = pd.read_csv(FILE_PATH)
# print no of rows and lows
print(df.shape)
```

```python
#know the structure
df.head()
```

```python
#create a folder called  dataset where we will place the covid images
# use os module to create a folder called covid and move covid images in to that folder


TARGET_DIR = "Dataset/Covid"
if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("Covid folder created")
```

```python
# check how many covid images are there we have 
cnt = 0
for (i,row) in df.iterrows():
    if row["finding"]=="COVID-19" and row["view"]=="PA":  #check for covid images and find the front view with elemenate(PA)
        filename = row["filename"]   # from metadat
        image_path = os.path.join(IMAGES_PATH, filename)
        image_copy_path = os.path.join(TARGET_DIR, filename)  # copy images from original to target dir
        shutil.copy2(image_path,image_copy_path)
       # print("Moving Image", cnt)
        cnt += 1
        
print(cnt)
```

```python
#sampling of images from Kaggle
import random
KAGGLE_FILE_PATH = "chest_xray_kaggle/train/NORMAL"
TARGET_NORMAL_DIR = "Dataset/Normal"
```
